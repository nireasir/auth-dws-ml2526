# Imports

```python
import gymnasium as gym
import renderlab as rl
import numpy as np
from stable_baselines3 import DQN
from stable_baselines3 import PPO
import os
```

# Ερώτημα 2

**1)** **reset()**: Xρησιμοποιείται για την αρχικοποίηση του περιβάλλοντος, δηλαδή επαναφέρει το περιβάλλον στην αρχική του κατάσταση.

**2)step(action)**: Δέχεται ως είσοδο μια ενέργεια (action) που επέλεξε ο πράκτορας και επιστρέφει τα εξής 5 αποτελέσματα:

*  **observation**: Η νέα κατάσταση του περιβάλλοντος.

*   **reward**: Την ανταμοιβή που κέρδισε ο πράκτορας από την ενέργεια.
*   **terminated**: true/false ανάλογα με το αν ο πράκτορας τερμάτισε.

*   **truncated**: true/false ανάλογα με το αν το επεισόδιο διακόπηκε.
*   **info**: Επιπλέον βοηθητικές πληροφορίες.

**3)render()**: Χρησιμοποιείται για οπτικοποίηση του περιβάλλοντος, ώστε να μπορούμε να δούμε τί κάνει ο πράκτορας.





# Ερώτημα 3

**1)Observation Space**: Διάνυσμα 8 διαστάσεων οι οποίες είναι: συντεταγμένες(x,y), οριζόντια ταχύτητα, κάθετη ταχύτητα, γωνία, γωνιακή ταχύτητα και δύο boolean τιμές που δείχνουν αν τα πόδια του σκάφους ακουμπούν στο έδαφος (μία για το αριστερό και μία για το δεξί).

2)**Action Space**: Εδώ υπάρχουν 4 πιθανές ενέργειες:


*  0: καμία ενέργεια

*  1: ενεργοποίηση αριστερού κινητήτα
*  2: ενεργοποίηση κεντρικού κινητήρα

*  3: ενεργοποίηση δεξιού κινητήρα

3)**Reward Function**: Στόχος είναι η ασφαλής προσγείωση σε συγκεκριμένο σημείο με όσο γίνεται λιγότερο καύσιμο. Η ανταμοιβή αυξάνεται όσο το σκάφος πλησιάζει στο σημείο και μειώνεται αν απομακρύνεται ή έχει μεγάλη ταχύτητα ενώ παίρνει και έξτρα πόντους για κάθε πόδι που ακουμπάει στο έδαφος. Η συντριβή δίνει -100 πόντους, ενώ η ασφαλής προσγείωση +100. Η χρήση κινητήρων μειώνει για λίγο την ανταμοιβή.



# Ερώτημα 4

```python
env = gym.make("LunarLander-v3", render_mode="rgb_array")
env=rl.RenderFrame(env,"./output")
observation,info=env.reset()
while True:
  action=env.action_space.sample()
  observation, reward, terminated, truncated, info =env.step(action)
  if terminated or truncated:
    break
env.play()
```

# Ερώτημα 5

```python
env = gym.make("LunarLander-v3", render_mode="rgb_array")
env=rl.RenderFrame(env,"./output")
total_rewards=[]
for episode in range(5):
  observation,info=env.reset()
  ep_reward=0
  while True:
    action=env.action_space.sample()
    observation, reward, terminated, truncated, info =env.step(action)
    ep_reward+=reward
    if terminated or truncated:
      break
  total_rewards.append(ep_reward)
avg_reward=sum(total_rewards)/5
print(f"average score={avg_reward}")
```

# Ερώτημα 6

```python
env_dqn = gym.make("LunarLander-v3", render_mode=None)
model_dqn=DQN("MlpPolicy",env_dqn,verbose=1)
model_dqn.learn(total_timesteps=70000)
model_dqn.save("dqn_lunar")
env_ppo = gym.make("LunarLander-v3", render_mode=None)
model_ppo=PPO("MlpPolicy",env_ppo,verbose=1)
model_ppo.learn(total_timesteps=70000)
model_ppo.save("ppo_lunar")
```
